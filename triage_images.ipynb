```python
import os
import re
import pandas as pd
```

```python
df = pd.read_csv('PokeDataset.csv')
trad = pd.read_csv('PokeTrad.csv')
```

```python
# make directory images/final/name for each pokemon
try:
    os.mkdir('images/final')
except FileExistsError:
    pass

for i in range(len(df)):
    name = df.iloc[i, 1].replace(':', '')
    try:
        os.mkdir('images/final/' + name)
    except FileExistsError:
        pass
```

folder 6 has white bg
folder 3 is converted from webp to png

```python
# iterate through images and move them to their respective folders
for folder in range(0, 17):
    if folder != 6:
        for file in os.listdir(f'images/pokemon{folder}/'):
            try:
                # extract pokemon number from file name
                num = re.findall(r'\d+', file)[0]
                # remove leading zeros
                num = int(num)
                # if num == 1:
                #     break
                # get pokemon name
                name = df.iloc[num-1, 1].replace(':', '')

                print(f"Moving images/pokemon{folder}/{file} to 'images/final/{name}/{str(len(os.listdir('images/final/' + name)))}.png")
                # copy file to images/final/name/count.png and count is the number of files in the folder
                os.rename(f"images/pokemon{folder}/{file}", f"images/final/{name}/{str(len(os.listdir('images/final/' + name)))}.png")
            except IndexError:
                # check if the file is the pokemon name in lowercase
                name = file.split('.')[0]
                # print()
                if name[:5] == 'mega-':
                    name = name[5:]
                elif name[:4] == 'tapu':
                    name = name.replace('-', ' ')
                # try to get first letter in uppercase
                try:
                    name = name[0].upper() + name[1:]
                    # remove after '-'
                    name = name.split('-')[0]
                    if name == 'Ho' or name == 'Hooh':
                        name = 'Ho-Oh'
                    elif name == 'M' or name == 'Mr':
                        name = 'Mr. Mime'
                    elif name == 'Lan' or name == 'Lan ':
                        name == "Escavalier" 
                    elif name == 'Mime':
                        name = 'Mime Jr'
                    elif name == 'Mmime':
                        continue
                    elif name == 'Nidoran':
                        continue
                    elif name == 'Porygonz':
                        name = 'Porygon-Z'
                    elif name == "Farfetchd":
                        name = "Farfetch'd"
                    elif name == 'Hakamo':
                        name = 'Hakamo-o'
                    elif name == 'Jangmo':
                        name = 'Jangmo-o'
                    elif name == 'Kommo':
                        name = 'Kommo-o'
                    elif name == 'Sirfetchd':
                        name = "Sirfetch'd"
                    elif name == 'Type':
                        continue


                    try :
                        name = trad[trad['French'] == name]['English'].values[0]
                    except IndexError:
                        print(f"Couldn't find {name} in trad table")
                    try :
                        # that means the file is the pokemon name so move it to its folder
                        # print(f"Moving images/pokemon{folder}/{file} to 'images/final/{name}/{str(len(os.listdir('images/final/' + name)))}.png")
                        os.rename(f'images/pokemon{folder}/{file}', f"images/final/{name}/{str(len(os.listdir('images/final/' + name)))}.png")
                    except FileNotFoundError:
                        print(f"Couldn't find {name} folder")
                except IndexError:
                    pass
```
